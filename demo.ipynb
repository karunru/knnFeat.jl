# iris dataset

```julia
using RDatasets
```

```julia
iris = dataset("datasets", "iris")
```

```julia
y = iris[:Species]
```

```julia
X = copy(iris)
delete!(X, :Species)
X = Array(X)
```

```julia
plot(x = X[:, 1], y = X[:, 2], color = y, Geom.point)
```

```julia
newX = knnExtract(X, y, 1, 5)
```

```julia
plot(x = newX[:, 1], y = newX[:, 2], color = y, Geom.point)
```
